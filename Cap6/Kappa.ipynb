```python
from statsmodels.stats.inter_rater import cohens_kappa, fleiss_kappa
import numpy as np
```

#Kappa de Cohen

```python
matrix = np.array([[22,3],
                   [5,20]])

print("Cohen's Kappa:", cohens_kappa(table = matrix))
```

#Kappa de Fleiss

```python
matrix = np.array([[3,0,0,0],
                   [1,2,0,0],
                   [0,0,3,0],
                   [3,0,0,0],
                   [0,3,0,0],
                   [0,3,0,0],
                   [0,0,0,3],
                   [1,2,0,0],
                   [0,0,0,3],
                   [0,0,0,3]])

print("Fleiss' Kappa:", fleiss_kappa(table = matrix, method = 'fleiss'))
```
