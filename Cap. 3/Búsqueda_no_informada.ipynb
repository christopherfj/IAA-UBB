```python
from collections import deque
import networkx as nx
from matplotlib import pylab
import random
import numpy as np
seed = 42
random.seed(seed)
np.random.seed(seed)
```

# Algoritmo BFS

```python
# BFS from given source s
def bfs(G, s):

    # Create a queue for BFS
    q = deque()

    # Initially mark all the vertices as not visited
    # When we push a vertex into the q, we mark it as
    # visited
    visited = [False] * len(G);

    # Mark the source node as visited and enqueue it
    visited[s] = True
    q.append(s)

    # Iterate over the queue
    while q:

        # Dequeue a vertex from queue and print it
        curr = q.popleft()
        print(curr, end=" ")

        # Get all adjacent vertices of the dequeued
        # vertex. If an adjacent has not been visited,
        # mark it visited and enqueue it
        for x in G[curr]:
            if not visited[x]:
                visited[x] = True
                q.append(x)
```

```python
# Create an empty graph
G = nx.Graph(seed=seed)

# Add edges to the graph based on the structure from the image
G.add_edges_from([ (0,3), (3,5), (0,2), (2,5), (2,4), (0,1) ])

# Spring layout positions nodes to reduce crossing edges
pos = nx.spring_layout(G, seed=seed)

fig = pylab.figure(1)
# Draw the graph using a layout that minimizes edge crossings
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=500, font_size=10, font_weight='bold', edge_color='gray')
pylab.title('BFS')
pylab.show()
```

```python
# Perform BFS traversal starting from vertex 0
print("BFS starting from 0: ")
s = 0
bfs(G, s)
```

#Algoritmo DFS

```python
# DFS from given source s
def dfs(G, s):

    # Create a stack for DFS
    stack = []

    # Initially mark all the vertices as not visited
    visited = [False] * len(G)

    # Mark the source node as visited and push it
    visited[s] = True
    stack.append(s)

    # Iterate over the stack
    while stack:

        # Pop a vertex from stack and print it
        curr = stack.pop()
        print(curr, end=" ")

        # Push all unvisited adjacent vertices
        for x in G[curr]:
            if not visited[x]:
                visited[x] = True
                stack.append(x)
```

```python
# Create an empty graph
G = nx.Graph(seed=seed)

# Add edges to the graph based on the structure from the image
G.add_edges_from([ (0,3), (3,5), (0,2), (2,5), (2,4), (0,1) ])

# Spring layout positions nodes to reduce crossing edges
pos = nx.spring_layout(G, seed=seed)

fig = pylab.figure(2)
# Draw the graph using a layout that minimizes edge crossings
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=500, font_size=10, font_weight='bold', edge_color='gray')
pylab.title('DFS')
pylab.show()
```

```python
s = 0
print("DFS from source:", s)
dfs(G, s)
```
