#Código CNN

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, Model, Input
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score

# =====================================================
# 1. Cargar datos
# =====================================================

(x, y), _ = keras.datasets.mnist.load_data()

x = x.astype("float32") / 255.0
x = np.expand_dims(x, axis=-1)

# Subset para acelerar
x = x[:5000]
y = y[:5000]

# =====================================================
# 2. Separar datos
# =====================================================

x_train, x_test, y_train, y_test = train_test_split(
    x,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y
)

y_train_oh = keras.utils.to_categorical(y_train, 10)
y_test_oh = keras.utils.to_categorical(y_test, 10)

# =====================================================
# 3. CNN original
# =====================================================

inputs = Input(shape=(28, 28, 1))

# Bloque 1
x = layers.Conv2D(32, 3, padding="same", use_bias=False)(inputs)
x = layers.BatchNormalization()(x)
x = layers.ReLU()(x)

x = layers.Conv2D(32, 3, padding="same", use_bias=False)(x)
x = layers.BatchNormalization()(x)
x = layers.ReLU()(x)

pool1 = layers.MaxPooling2D()(x)

# Bloque 2
x = layers.Conv2D(64, 3, padding="same", use_bias=False)(pool1)
x = layers.BatchNormalization()(x)
x = layers.ReLU()(x)

x = layers.Conv2D(64, 3, padding="same", use_bias=False)(x)
x = layers.BatchNormalization()(x)
x = layers.ReLU()(x)

pool2 = layers.MaxPooling2D()(x)

# Bloque 3
x = layers.Conv2D(128, 3, padding="same", use_bias=False)(pool2)
x = layers.BatchNormalization()(x)
x = layers.ReLU()(x)

x = layers.Conv2D(128, 3, padding="same", use_bias=False)(x)
x = layers.BatchNormalization()(x)
x = layers.ReLU()(x)

pool3 = layers.MaxPooling2D()(x)

x = layers.Flatten()(pool3)
x = layers.Dropout(0.4)(x)

x = layers.Dense(256, activation="relu")(x)
outputs = layers.Dense(10, activation="softmax")(x)

model = Model(inputs, outputs)

model.summary()

# =====================================================
# 4. Compilar
# =====================================================

model.compile(
    optimizer="adam",
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

# =====================================================
# 5. Entrenar
# =====================================================

history = model.fit(
    x_train,
    y_train_oh,
    epochs=10,
    batch_size=64,
    validation_split=0.2,
    verbose=1
)

# =====================================================
# 6. Predicciones
# =====================================================

pred = model.predict(x_test, verbose=0)

y_pred = np.argmax(pred, axis=1)

# =====================================================
# 7. F1
# =====================================================

f1 = f1_score(y_test, y_pred, average="macro")

print(f"\nF1-macro: {f1:.4f}")

# =====================================================
# 8. Curvas de entrenamiento
# =====================================================

plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
plt.plot(history.history["loss"], label="Train")
plt.plot(history.history["val_loss"], label="Val")
plt.title("Loss")
plt.legend()
plt.grid()

plt.subplot(1,2,2)
plt.plot(history.history["accuracy"], label="Train")
plt.plot(history.history["val_accuracy"], label="Val")
plt.title("Accuracy")
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()

# =====================================================
# 9. Feature maps
# =====================================================

feature_model = Model(
    inputs=model.input,
    outputs=[pool1, pool2, pool3]
)

features = feature_model.predict(x_test[0:1], verbose=0)

plt.figure(figsize=(12,6))

for i, fmap in enumerate(features):
    for j in range(6):

        plt.subplot(3,6,i*6+j+1)

        plt.imshow(
            fmap[0, :, :, j],
            cmap="viridis"
        )

        plt.axis("off")

        if j == 0:
            plt.ylabel(f"Bloque {i+1}")

plt.suptitle("Mapas de características")
plt.tight_layout()
plt.show()

# =====================================================
# 10. Predicciones
# =====================================================

plt.figure(figsize=(10,4))

for i in range(10):

    plt.subplot(2,5,i+1)

    plt.imshow(
        x_test[i].squeeze(),
        cmap="gray"
    )

    plt.axis("off")

    plt.title(
        f"R:{y_test[i]}\nP:{y_pred[i]}"
    )

plt.tight_layout()
plt.show()
```
