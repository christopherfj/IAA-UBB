```python
class Persona:
    def __init__(self, nombre, edad=0):
        self.nombre = nombre
        self.edad = edad
    def hablar(self):
        print("Hola, me llamo", self.nombre)
    def __add__(self, otra):
        return self.edad + otra.edad

pedro = Persona("Pedro", 10)
#pedro.hablar()
#print(pedro.edad)

juan = Persona("Juan", 30)

pedro+juan
```
